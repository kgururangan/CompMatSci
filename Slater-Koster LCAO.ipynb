In this script, we will solve for the band structures of various 2D and 3D crystals using the linear combination of atomic orbitals (LCAO) approach under the tight-binding approximation. The main idea in LCAO is to expand the many-body wavefunction $\psi$ in a linear combination of wavefunctions belonging to each atom/nucleus in the system:

$$\psi = \sum_{A=1}^{N_{nuc}}c^A\phi_{A} = c^A\phi_{A} + c^B\phi_{B} + c^C\phi_{C} + \ldots$$

Each atomic wavefunction is itself expanded in an orthonormal basis $\{\chi_j(\textbf{r})\}$ for $j=1\ldots N_{orb}$ centered at the atom position $\textbf{R}_A$. Note that $\langle\chi^{A}_i|\chi^{A}_j\rangle = \delta_{ij}$ but  $\langle\chi^{A}_i|\chi^{B}_j\rangle \neq \delta_{ij}$ i.e. the basis is only orthogonal when centered at the same nucleus. This condition is easily accomplished using linear combinations of localized hydrogenic orbitals for the $\chi_j(\textbf{r})$. If we start with the eigenfunctions of the Coulomb potential:

$$|nlm\rangle = R_{nl}(r)Y_l^m(\theta,\phi)$$  

$$
\begin{cases}
|200\rangle = \dfrac{1}{2\sqrt{2\pi}}\left(\dfrac{Z}{a_\mu}\right)^{3/2}\left(1-\dfrac{Zr}{2a_\mu}\right)\exp\left(\frac{-Zr}{a_\mu}\right)\\
|210\rangle = \dfrac{1}{4\sqrt{2\pi}}\left(\dfrac{Z}{a_\mu}\right)^{3/2}\dfrac{Zr}{a_\mu}\exp\left(-\frac{Zr}{2a_\mu}\right)\cos\theta\\
|211\rangle = -\dfrac{1}{8\sqrt{\pi}}\left(\dfrac{Z}{a_\mu}\right)^{3/2}\dfrac{Zr}{a_\mu}\exp\left(-\frac{Zr}{2a_\mu}\right)\sin\theta e^{i\phi}\\
|21 \text{-}1\rangle = \dfrac{1}{8\sqrt{\pi}}\left(\dfrac{Z}{a_\mu}\right)^{3/2}\dfrac{Zr}{a_\mu}\exp\left(-\frac{Zr}{2a_\mu}\right)\sin\theta e^{-i\phi}
\end{cases}
$$

The LCAO basis for sp-valent materials are the $s$, $p_x$, $p_y$, and $p_z$ orbitals defined by the following linear combinations:

$$\chi_j(\textbf{r}) = 
\begin{cases}
\chi_{2s}(\textbf{r}) = |200\rangle \\
\chi_{2p_z}(\textbf{r}) = |210\rangle \\
\chi_{2p_x}(\textbf{r}) = -\frac{1}{\sqrt{2}}(|211\rangle - |21\text{-}1\rangle) \\
\chi_{2p_y}(\textbf{r}) = \frac{i}{\sqrt{2}}(|211\rangle + |21\text{-}1\rangle)
\end{cases}
$$

In general, the LCAO wavefunctions takes the form:

$$\phi_{A} = \sum_{j=1}^{N_{orb}} c^{A}_j \chi^{A}_j(\textbf{r}) = \sum_{j=1}^{N_{orb}} c^{A}_j \chi_j(\textbf{r}-\textbf{R}_A)$$

$$\psi(\textbf{r}) = \sum_{A=1}^{N_{nuc}}\sum_{j=1}^{N_{orb}} c^{A}_j\chi^A_j(\textbf{r}) = \sum_{A=1}^{N_{nuc}}\sum_{j=1}^{N_{orb}}c^A_j\chi_j(\textbf{r}-\textbf{R}_A)$$

With this expansion, the Schrodinger equation can be converted into matrix form by taking $\left\langle\chi^B_k\right|$ and summing over all $B$ and $k$ to produce the eigenvalue equation:

$$\sum_{A=1}^{N_{nuc}}\sum_{j=1}^{N_{orb}}\left( H\left|\chi^A_j\right\rangle - E\left|\chi^A_j\right\rangle \right)c^A_j = 0$$

$$\sum_{B=1}^{N_{nuc}}\sum_{k=1}^{N_{orb}}\sum_{A=1}^{N_{nuc}}\sum_{j=1}^{N_{orb}}\left( \left\langle\chi^B_k\right|\hat{H}\left|\chi^A_j\right\rangle - E\left\langle\chi^B_k\right|\chi^A_j\rangle\right)c^A_{j} = 0$$

Typically, we arrange the coefficients $c^A_j$ in a single column vector (this is equivalent to considering the composite label mapping $(j,A)\rightarrow J$) such that 

$$
\textbf{c}=
  \begin{bmatrix}
    c^A_1\\
    \vdots\\
    c^A_{N_{orb}}\\
    \vdots\\
    c^{N_{nuc}}_1\\
    \vdots\\
    c^{N_{nuc}}_{N_{orb}}
  \end{bmatrix}
$$

The composite label runs from $J = 1\ldots N_{bands}$ where $N_{bands}=N_{orb}N_{nuc}$

$$\sum_{I=1}^{N_{bands}}\sum_{J=1}^{N_{bands}}\left(H_{IJ}-ES_{IJ}\right)c_J = 0\Rightarrow \left(\textbf{H} - E\textbf{S}\right)\textbf{c} = \textbf{0}$$

In this form, the Hamiltonian and overlap matrices are organized into blocks according to each nucleus featured in the problem:

$$
\textbf{H} = 
\begin{bmatrix}
    \textbf{H}_A & \textbf{H}_{AB} & \ldots & \textbf{H}_{A N_{nuc}} \\
    \textbf{H}_{BA} & \textbf{H}_{B} & \ldots & \textbf{H}_{B N_{nuc}} \\
    \vdots & \vdots & \ddots & \vdots \\
    \textbf{H}_{N_{nuc} A} & \textbf{H}_{N_{nuc} B} & \ldots & \textbf{H}_{N_{nuc}}
\end{bmatrix}
$$

$$
\textbf{S} = 
\begin{bmatrix}
    \textbf{I} & \textbf{S}_{AB} & \ldots & \textbf{S}_{A N_{nuc}} \\
    \textbf{S}_{BA} & \textbf{I} & \ldots & \textbf{S}_{B N_{nuc}} \\
    \vdots & \vdots & \ddots & \vdots \\
    \textbf{S}_{N_{nuc} A} & \textbf{S}_{N_{nuc} B} & \ldots & \textbf{I}
\end{bmatrix}
$$

Note that $\textbf{H}_{BA} = \textbf{H}^\dagger_{AB}$ and $S_{JI} = (-1)^p{S}_{IJ}$ where $p$ is the sum of the parities of the orbitals $\chi_I$ and $\chi_J$ ($p = 0,1$ for even and odd orbitals with respect to inversion, respectively).


In the following, we will make the orthogonal orbital assumption which sets $\textbf{S} = \textbf{I}$. Of course, we know this is wrong by definition of the $\chi_j$, but it lets us solve a simpler eigenvalue problem as opposed to a generalized eigenvalue problem. Relaxing this approximation is easy to handle as the overlap matrix defined in the function "SKmat."The goal is simply to diagonalize $\textbf{H}$ to obtain the eigenvalues and eigenvectors by solving $\det\left(\textbf{H}-E\textbf{I}\right) = 0$. At the moment, this formalism is quite general and only hinges on the the suitability of $\{\chi_J\}$ as a basis to describe the problem. The major approximation moving forward is to define the action of the Hamiltonian $H$ on a specific state using the tight-binding approximation. This approximation results in a hopping matrix $t_{IJ}$ used to approximate the kinetic energy in lattice systems. It can be written very compactly using second-quantized fermion operators:

$$H = \sum_{IJ}^{NN}t_{IJ}c^\dagger_Jc_I + \sum_{IJ}^{NNN}t_{IJ}c^\dagger_Jc_I + \sum_{IJK}^{3-atom}\left[t^{(1)}_{IJK}(c^\dagger_Ic_Jc_K + c_Ic^\dagger_Jc_K + c_Ic_Jc^\dagger_K) + t^{(2)}_{IJK}(c^\dagger_Ic^\dagger_Jc_K + c_Ic^\dagger_Jc^\dagger_K + c^\dagger_Ic_Jc^\dagger_K)\right] + \ldots$$

where $c^\dagger_J$ and $c_J$ are creation and annihilation operators, respectively, for the orbital $\chi_J(\textbf{r})$. The sums are taken over all nearest-neighbors up to an arbitrary number of shells, including multi-atom clusters and higher. The lattice topology will dictate the number and distance of nearest-neighbors for the Hamiltonian. For these problems, we will not take spin into account, but this formulation is easily generalized to include $c^\dagger_J\rightarrow c^\dagger_{J\sigma}$, resulting in the well-known Hubbard-like Hamiltonian. The hopping amplitude to go from orbital $I$ to $J$ is approximated as the overlap between the orbitals, therefore

$$t_{IJ} = \left\langle\chi_I\right|\chi_J\rangle$$

Because these hopping amplitudes correspond to the overlaps between s-, p-, and d-orbitals, tight-binding models typically give these values physically intuitive labels based on the types of overlapping orbitals and the $\sigma$- and $\pi$-character of the bond. These 2-atom center integrals are called Slater-Koster parameters and are labelled $\{ss\sigma(d), pp\sigma(d), sp\sigma(d), pp\pi(d), sd\sigma(d), pd\sigma(d), pd\pi(d), dd\delta(d) \}$ where $d$ is the distance between the site centers. For two arbitrarily oriented orbitals in 3D space, the general interaction can be written as a linear combination of Slater-Koster parameters via projection onto the bond and normal axes.   

To apply tight-binding to solids, one must place the nuclei at points located by the vectors $\{\textbf{R}_{uvw} = u\textbf{a}_1 + v\textbf{a}_2 + \textbf{a}_3\}$ for $u,v,w = 0\ldots L-1$ to form an L-periodic crystal lattice, effectively relabelling the nuclear index on the orbitals from $\chi_j^A(\textbf{r})\rightarrow \chi^{uvw}_j(\textbf{r})$. Using the projection operator method from group theory, one can show that the Bloch-symmetrized form of any orbital is 

$$\rho^{(uvw)}\chi^{000}_j(\textbf{r}) = \sum_{u,v,w=0}^{L-1}\left[C^{uvw}\right]^* P^{uvw}\chi^{000}_j(\textbf{r})= \dfrac{1}{L^{3/2}}\sum_{u,v,w=0}^{L-1}e^{\frac{-2\pi i (u+v+w)}{L}}\left| \chi^{uvw}_j \right\rangle$$
$$\equiv \dfrac{1}{L^{3/2}}\sum_{u,v,w=0}^{L-1}e^{-i\textbf{k}\cdot\textbf{R}_{uvw}}\left| \chi^{uvw}_j \right\rangle$$

where $\textbf{k} = \frac{\mu}{L}\textbf{b}_1 + \frac{\nu}{L}\textbf{b}_2 + \frac{\lambda}{L}\textbf{b}_3$ for $\mu,\nu,\lambda = 0\ldots L-1$ defines the wavevector, $C^{uvw}$ is the character of the $uvw^{\mathrm{th}}$ symmetry operation, and $P^{uvw}$ is the operator that translates a function by $\textbf{R}_{uvw}$. The final result can be simply interpreted as Bloch's theorem. It is also easy to show that this is the only unique orbital up to a phase factor, therefore, by symmetrization with respect to the lattice, we need not separately consider the sites related by symmetry (for which there are formally an infinite number of as $L\rightarrow\infty$). We have effectively traded an infinite-dimensional problem Hamiltonian for a finite one that parametrically depends on a symmetry wavevector $\textbf{k}$. Therefore, the general form for the hopping matrix element becomes

$$t_{ij} = \dfrac{1}{L^3}\sum_{u'v'w' = 0}^{L-1}\sum_{u,v,w=0}^{L-1} e^{-i\textbf{k}\cdot(\textbf{R}_{uvw}-\textbf{R}_{u'v'w'})}\langle \chi^{u'v'w'}_i | \chi^{uvw}_j \rangle$$

The term $\langle \chi^{u'v'w'}_i | \chi^{uvw}_j \rangle$ is simply the overlap between hydrogenic orbitals located at sites $\textbf{R}_{uvw}$ and $\textbf{R}_{u'v'w'}$ which is expressable using projections onto Slater-Koster interactions. The neighbor vector within a certain shell $\textbf{d}_i = \textbf{R}_{uvw} - \textbf{R}_{u'v'w'}$ defines the phase factor that is picked up. In tight-binding literature, this expression is often written as 

$$t_{ij} = \sum_{i=1}^{NN}\gamma_j(|\textbf{d}_i|)e^{-i\textbf{k}\cdot\textbf{d}_i}$$ 

which is often called a Bloch sum. It adds up the phase factor contributions due to (nearest) neighbors weighted by the tight-binding parameter $\gamma$ which is a function of site separation. In the above, we can see that the tight-binding parameters are simply orbital overlaps. As an example, consider a 1D s-valent chain with nearest-neighbors at $\pm a_0\textbf{E}_x$. Given that the orbital overlap between s-orbtials is simply $ss\sigma(a_0)$, the Bloch sum evaluates to 

$$ss\sigma(a_0)(e^{i\textbf{k}\cdot a_0\textbf{E}_x} + e^{-i\textbf{k}\cdot a_0\textbf{E}_x}) = 2 ss\sigma(a_0)\cos(k_xa_0)$$

which is the simplest example of an energy band. The Bloch sum and resulting energy bands are entirely structure-dependent and lies at the heart of the relationship between topology and electronic structure.

At this point, it is important to understand what is not captured in tight-binding models. The simple Hamiltonian only crudely approximates the kinetic energy, neglecting Coulomb interactions (Hartree energy), external potentials, and exchange or correlation effects. 

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import math
import cmath
pi = math.pi
cm = [0,0,0]
#cm =[0.3, 0.4, 0.8]
```

```python
# Define orbital energies and interaction parameters for carbon
aNN = 1.42
es = 0
ep = 0
ed = 0
sss_b = -5.00
sps_b = 4.70
pps_b = 5.50
ppp_b = -1.55
sds_b = 1
pds_b = -1
pdp_b = -1
dds_b = -2
ddp_b = 1.5
ddd_b = -1

def beta(r):
    r0 = 1.536
    rc = 2.18
    nc = 6.5
    n = 2.0
    return (r0/r)**n*np.exp(-n*(-(r/rc)**nc + (r0/rc)**nc))

r = beta(aNN)

sss = sss_b*r
sps = sps_b*r
pps = pps_b*r
ppp = ppp_b*r
sds = sds_b*r
pds = pds_b*r
pdp = pdp_b*r
dds = dds_b*r
ddp = ddp_b*r
ddd = ddd_b*r

# From Popov et al
es = 0
ep = 0
sss = -4.80
sps = 4.75
pps = 4.39
ppp = -2.56
```

```python
def SKparam(d):
    R = np.linalg.norm(d)
    bA = 0.529
    sss = (1 + 5.158/bA*R + 0.66/(bA**2)*R**2 - 0.0815/bA**3 * R**3)*np.exp(-(1.08/math.sqrt(bA))**2*R)
    sps = (8.873/bA*R - 16.24/bA**2*R**2 + 5.182/bA**3*R**3)*np.exp(-(1.24/math.sqrt(bA))**2*R)
    pps = (1 + 11.25/bA*R - 1.17/bA**2*R**2 - 1.06/bA**3*R**3)*np.exp(-(1.14/math.sqrt(bA))**2*R)
    ppp = (1 - 692.18/bA*R + 396.15/bA**2*R**2 - 13.82/bA**3*R**3)*np.exp(-(1.572/math.sqrt(bA))**2*R)
    
    return sss,sps,pps,ppp

def SKmat(d):
    # Unit vector between atom centers
    dn = d/np.linalg.norm(d)
    # Direction cosines
    l1 = np.dot(dn,np.array([1,0,0]))
    l2 = np.dot(dn,np.array([0,1,0]))
    l3 = np.dot(dn,np.array([0,0,1]))
    
    #sss,sps,pps,ppp = SKparam(d)
    
    # Slater-Koster parameter table
    S = np.zeros([9,9])
    
    # s-orbital
    S[0,0] = sss
    S[0,1] = l1*sps; S[1,0] = -S[0,1];
    S[0,2] = l2*sps; S[2,0] = -S[0,2];
    S[0,3] = l3*sps; S[3,0] = -S[0,3];
    S[0,4] = math.sqrt(3)*l1*l2*sds; S[4,0] = S[0,4];
    S[0,5] = math.sqrt(3)*l2*l3*sds; S[5,0] = S[0,5];
    S[0,6] = math.sqrt(3)*l3*l1*sds; S[6,0] = S[0,6];
    S[0,7] = 0.5*math.sqrt(3)*(l1**2-l2**2)*sds; S[7,0] = S[0,7];
    S[0,8] = (l3**2 - 0.5*(l1**2+l2**2))*sds; S[8,0] = S[0,8];
    # px-orbital
    S[1,1] = l1**2*pps + (1-l1**2)*ppp; 
    S[1,2] = l1*l2*pps - l1*l2*ppp; S[2,1] = S[1,2];
    S[1,3] = l1*l3*pps - l1*l3*ppp; S[3,1] = S[1,3];
    S[1,4] = math.sqrt(3)*l1**2*l2*pds + l2*(1-2*l1**2)*pdp; S[4,1] = -S[1,4];
    S[1,5] = math.sqrt(3)*l1*l2*l3*pds - 2*l1*l2*l3*pdp; S[5,1] = -S[1,5];
    S[1,6] = math.sqrt(3)*l1**2*l3*pds + l3*(1-2*l1**2)*pdp; S[6,1] = -S[1,6];
    S[1,7] = 0.5*math.sqrt(3)*l1*(l1**2-l2**2)*pds + l1*(1-l1**2 + l2**2)*pdp; S[7,1] = -S[1,7];
    S[1,8] = l1*(l3**2-0.5*(l1**2+l2**2))*pds - math.sqrt(3)*l1*l3**2*pdp; S[8,1] = -S[1,8];
    # py-orbital
    S[2,2] = l2**2*pps + (1-l2**2)*ppp;
    S[2,3] = l2*l3*pps - l2*l3*ppp; S[3,2] = S[2,3];
    S[2,4] = math.sqrt(3)*l2**2*l1*pds + l1*(1-2*l2**2)*pdp; S[4,2] = -S[2,4];
    S[2,5] = math.sqrt(3)*l2**2*l3*pds + l3*(1-2*l2**2)*pdp; S[5,2] = -S[2,5];
    S[2,6] = math.sqrt(3)*l1*l2*l3*pds - 2*l1*l2*l3*pdp; S[6,2] = -S[2,6];
    S[2,7] = 0.5*math.sqrt(3)*l2*(l1**2-l2**2)*pds - l2*(1+l1**2-l2**2)*pdp; S[7,2] = -S[2,7];
    S[2,8] = l2*(l3**2 - 0.5*(l1**2+l2**2))*pds - math.sqrt(3)*l2*l3**2*pdp; S[8,2] = -S[2,8];
    # pz-orbital
    S[3,3] = l3**2*ppp + (1-l3**2)*ppp;
    S[3,4] = math.sqrt(3)*l1*l2*l3*pds - 2*l1*l2*l3*pdp; S[4,3] = -S[3,4];
    S[3,5] = math.sqrt(3)*l2**2*l1*pds + l1*(1-2*l2**2)*pdp; S[5,3] = -S[3,5];
    S[3,6] = math.sqrt(3)*l3**2*l1*pds + l1*(1-2*l3**2)*pdp; S[6,3] = -S[3,6];
    S[3,7] = 0.5*math.sqrt(3)*l3*(l1**2 - l2**2)*pds - l3*(l1**2 - l2**2)*pdp; S[7,3] = -S[3,7];
    S[3,8] = l3*(l3**2-0.5*(l1**2+l2**2))*pds + math.sqrt(3)*l3*(l1**2+l2**2)*pdp; S[8,3] = -S[3,8];
    # dxy-orbital
    S[4,4] = 3*l1**2*l2**2*dds + (l1**2+l2**2-4*l1**2*l2**2)*ddp + (l3**2+l1**2*l2**2)*ddd;
    S[4,5] = 3*l1*l2**2*l3*dds + l1*l3*(1-4*l2**2)*ddp + l1*l3*(l2**2 - 1)*ddd; S[5,4] = S[4,5];
    S[4,6] = 3*l1**2*l2*l3*dds + l2*l3*(1-4*l1**2)*ddp + l2*l3*(l1**2-1)*ddd; S[6,4] = S[4,6];
    S[4,7] = 1.5*l1*l2*(l1**2-l2**2)*dds + 2*l1*l2*(l2**2-l1**2)*ddp + 0.5*l1*l2*(l1**2-l2**2)*ddd; S[7,4] = S[4,7];
    S[4,8] = math.sqrt(3)*l1*l2*(l3**2-0.5*(l1**2+l2**2))*dds - 2*math.sqrt(3)*l1*l2*l3**2*ddp + 0.5*math.sqrt(3)*l1*l2*(1+l3**2)*ddd; S[8,4] = S[4,8];
    # dyz-orbital
    S[5,5] = 3*l2**2*l3**2*dds + (l2**2+l3**2-4*l2**2*l3**2)*ddp + (l1**2+l2**2*l3**2)*ddd
    S[5,6] = 3*l2*l3**2*l1*dds + l2*l1*(1-4*l3**2)*ddp + l2*l1*(l3**2 - 1)*ddd; S[6,5] = S[5,6];
    S[5,7] = 1.5*l2*l3*(l1**2-l2**2)*dds-l2*l3*(1+2*(l1**2-l2**2))*ddp+l2*l3*(1+0.5*(l1**2-l2**2))*ddd; S[7,5] = S[5,7];
    S[5,8] = math.sqrt(3)*l2*l3*(l3**2-0.5*(l1**2+l2**2))*dds+math.sqrt(3)*l2*l3*(l1**2+l2**2-l3**2)*ddp-0.5*math.sqrt(3)*l2*l3*(l1**2+l2**2)*ddd; S[8,5] = S[5,8];
    # dxz-orbital
    S[6,6] = 3*l3**2*l1**2*dds + (l3**2+l1**2-4*l3**2*l1**2)*ddp + (l2**2+l3**2*l1**2)*ddd;
    S[6,7] = 1.5*l3*l1*(l1**2-l2**2)*ddp+l3*l1*(1-2*(l1**2-l2**2))*ddp-l3*l1*(1-0.5*(l1**2-l2**2))*ddd; S[7,6] = S[6,7];
    S[6,8] = math.sqrt(3)*l1*l3*(l3**2-0.5*(l1**2+l2**2))*dds + math.sqrt(3)*l1*l3*(l1**2+l2**2-l3**2)*ddp - 0.5*math.sqrt(3)*l1*l3*(l1**2+l2**2)*ddd; S[8,6] = S[6,8];
    # d(x^2-y^2)-orbital
    S[7,7] = 0.75*(l1**2-l2**2)**2*dds+(l1**2+l2**2-(l1**2-l2**2)**2)*ddp+(l3**2+0.25*(l1**2-l2**2)**2)*ddd
    S[7,8] = 0.5*math.sqrt(3)*(l1**2-l2**2)*(l3**2-0.5*(l1**2+l2**2))*dds+math.sqrt(3)*l3**2*(l2**2-l1**2)*ddp+0.25*math.sqrt(3)*(1+l3**2)*(l1**2-l2**2)*ddd; S[8,7] = S[7,8];
    # d(3z^2-r^2)-orbital
    S[8,8] = (l3**2-0.5*(l1**2+l2**2))**2*dds+3*l3**2*(l1**2+l2**2)*ddp+0.75*(l1**2+l2**2)**2*ddd
    
    return S

def tmatrix(k,NLIST,orbitals):
    if orbitals == 'spd':
        A = np.zeros([9,9],dtype=complex)  
        for i in range(NLIST.shape[0]):
            delta = NLIST[i,:]
            SK = SKmat(delta)
            A = A + np.exp(-1j*np.dot(k,delta))*SK
        return A + np.diag([es,ep,ep,ep,ed,ed,ed,ed,ed])
    elif orbitals == 'sp':
        A = np.zeros([4,4],dtype=complex)  
        for i in range(NLIST.shape[0]):
            delta = NLIST[i,:]
            SK = SKmat(delta); SK = SK[0:4,0:4];
            A = A + np.exp(-1j*np.dot(k,delta))*SK;
        return A + np.diag([es,ep,ep,ep])
    else:
        A = np.zeros([1,1],dtype=complex)
        for i in range(NLIST.shape[0]):
            delta = NLIST[i,:]
            SK = SKmat(delta); SK = SK[0,0];
            A = A + np.exp(-1j*np.dot(k,delta))*SK;
        return A + es
    
def kPathVec(coords,nkpt):
    
    npaths = coords.shape[0] - 1
    
    d = np.zeros(npaths)
    kx = np.empty(0)
    ky = np.empty(0)
    kz = np.empty(0)
    kplot = np.empty(0)
    
    for i in range(npaths):
        d[i] = np.linalg.norm(coords[i+1,:] - coords[i,:])
        A = coords[i,0] + (coords[i+1,0] - coords[i,0])*np.linspace(0,1,nkpt)
        B = coords[i,1] + (coords[i+1,1] - coords[i,1])*np.linspace(0,1,nkpt)
        C = coords[i,2] + (coords[i+1,2] - coords[i,2])*np.linspace(0,1,nkpt)
        kx = np.concatenate( (kx, A) , axis=0 )
        ky = np.concatenate( (ky, B) , axis=0 )
        kz = np.concatenate( (kz, C) , axis=0 )
        if i == 0:
            kplot = np.concatenate( (kplot, np.linspace(0,d[0],nkpt)), axis = 0) 
        elif i == npaths:    
            kplot = np.concatenate( (kplot, np.linspace(np.sum(d[0:i]),np.sum(d[0:-1]),nkpt) ), axis = 0 )
        else:
            kplot = np.concatenate( (kplot, np.linspace(np.sum(d[0:i]),np.sum(d[0:i+1]),nkpt) ), axis = 0 )
    kvec = np.vstack( (kx,ky,kz) )
    
    return kvec, kplot


def stateCount( E, Elow, Ehigh ):
# Counts number of states in interval Ehigh - Elow in vector of energies E
    i_start = np.zeros([len(E)]);
    i_end = np.zeros([len(E)]);
    dE = np.abs(Ehigh - Elow)
    for i in range(len(E)):   
        if (np.abs(E[i] - Elow) < dE):     
            i_start[i] = i;  
        if (np.abs(E[i] - Ehigh) < dE):
            i_end[i] = i;
    Nstates = np.sum(np.abs(i_end - i_start))
    return Nstates
```

```python
# (1) 2D Square Lattice
Natoms = 1
orbitals = 'spd'

if orbitals == 'sp':
    Norb = 4
if orbitals == 'spd':
    Norb = 9
if orbitals == 's':
    Norb = 1
    
numbands = Natoms*Norb

aNN = 1.42;
a1 = aNN*np.array([1, 0, 0])
a2 = aNN*np.array([0, 1, 0])
b1 = 2*math.pi/aNN*np.array([1,0, 0])
b2 = 2*math.pi/aNN*np.array([0,1, 0])

A = np.vstack( (a1, a2) )
B = np.vstack( (b1, b2) )

# Define list of neighbors
NLIST = np.vstack( ([a1],[-a1],[a2],[-a2]) )

kG = np.array([0,0,0])
kX = math.pi/(aNN)*np.array([1,0,0])
kM = math.pi/(aNN)*np.array([1,1,0])

coords = np.vstack( (kG, kX, kM, kG) )
nkpt = 200

kvec, kplot = kPathVec(coords,nkpt)

# Calculate and plot bandstructure along specific path

E = np.zeros([numbands,len(kplot)])

for i in range(len(kplot)):
    Hmat = tmatrix(kvec[:,i],NLIST,orbitals)
    e,v = np.linalg.eigh(Hmat)
    E[:,i] = e
    
f = plt.figure(figsize=(18, 10)); 
for i in range(numbands):
    plt.scatter(kplot,E[i,:],c=cm,s=3)
    plt.xlabel(r'$k$',fontsize=20)
    plt.ylabel('Energy [eV]',fontsize=20)
    plt.title('Band Structure of spd-valent 2D Square Crystal',fontsize=20)
    plt.xticks([kplot[0],kplot[nkpt-1],kplot[2*nkpt-1],kplot[3*nkpt-1]],(r"$\Gamma$","X","M",r"$\Gamma$"),fontsize=20)
    plt.yticks(fontsize=20)
    plt.grid(True)
    plt.xlim([kplot[0],kplot[-1]])
    
# Density of states
    
dE = 0.1
L = 100

ksamp = np.zeros([3,L**2])
Esamp = np.zeros([numbands,L**2])

count = 0
for i in range(L):
    for j in range(L):
        ksamp[:,count] = i/L*b1 + j/L*b2
        count = count + 1

for i in range(ksamp.shape[1]):   
    Hmat = tmatrix(ksamp[:,i],NLIST,orbitals)
    e,v = np.linalg.eigh(Hmat)
    Esamp[:,i] = e
    
Emin = np.min(np.min(Esamp))
Emax = np.max(np.max(Esamp))
Esteps = np.arange(Emin,Emax,dE)

DOS = np.zeros([len(Esteps)])
for i in range(len(Esteps)-1): 
    for j in range(numbands):
        DOS[i] = DOS[i] + 1/dE * ( stateCount(Esamp[j,:],Esteps[i],Esteps[i+1]) )

g = plt.figure(figsize=(18, 10)); 
plt.plot(Esteps,DOS,color=cm)
plt.xlabel('Energy [eV]',fontsize=20)
plt.ylabel('# States/eV',fontsize=20)
plt.title('Density of States',fontsize=20)
plt.grid(True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axis('tight')
```

```python
# (2) 2D Honeycomb Lattice

Natoms = 2
orbitals = 'sp'

if orbitals == 'sp':
    Norb = 4
if orbitals == 'spd':
    Norb = 9
if orbitals == 's':
    Norb = 1
    
numbands = Natoms*Norb

next_nearest = 0

aNN = 1.42;
a1 = aNN/2*np.array([3, math.sqrt(3), 0])
a2 = aNN/2*np.array([3,-math.sqrt(3), 0])
b1 = 2*math.pi/(3*aNN)*np.array([1,math.sqrt(3), 0])
b2 = 2*math.pi/(3*aNN)*np.array([1,-math.sqrt(3), 0])

A = np.vstack( (a1, a2) )
B = np.vstack( (b1, b2) )

# Define list of neighbors
delta1 = aNN/2*np.array([1,math.sqrt(3),0])
delta2 = aNN/2*np.array([1,-math.sqrt(3),0])
delta3 = -aNN*np.array([1,0,0])
NLIST = np.vstack( ([delta1],[delta2],[delta3]) )
NNLIST = np.vstack( ([a1],[-a1],[a2],[-a2],[a2-a1],[a1-a2]) )

kG = np.array([0,0,0])
kK = 2*pi/(3*aNN)*np.array([1,1/math.sqrt(3),0])
kM = 2*pi/(3*aNN)*np.array([1,0,0])
kKp = 2*pi/(3*aNN)*np.array([1,-1/math.sqrt(3),0])

coords = np.vstack( (kG, kM, kK, kG) )
nkpt = 200

kvec, kplot = kPathVec(coords,nkpt)

# Calculate and plot bandstructure along specific path

E = np.zeros([numbands,len(kplot)])

for i in range(len(kplot)):
    Hmat = np.zeros([numbands,numbands])
    for j1 in range(Natoms):
        for j2 in range(j1,Natoms):
            kernel = np.zeros([Natoms,Natoms])
            kernel[j1,j2] = 1
            if j1 != j2:
                Hab = tmatrix(kvec[:,i],NLIST,orbitals)
                Hba = Hab.conj().transpose()
                Hmat = Hmat + np.kron(kernel,Hab) + np.kron(kernel.transpose(),Hba)
            else:
                if next_nearest == 1:
                    Ha = tmatrix(kvec[:,i],NNLIST,orbitals)
                else:
                    Ha = np.diag([es,ep,ep,ep])
                Hmat = Hmat + np.kron(kernel,Ha)
    e,v = np.linalg.eigh(Hmat)
    E[:,i] = e
    
f = plt.figure(figsize=(18, 10)); 
for i in range(numbands):
    plt.scatter(kplot,E[i,:],c=cm,s=3)
    plt.xlabel(r'$k$',fontsize=20)
    plt.ylabel('Energy [eV]',fontsize=20)
    plt.title('Band Structure of sp-valent 2D Honeycomb Crystal',fontsize=20)
    plt.xticks([kplot[0],kplot[nkpt-1],kplot[2*nkpt-1],kplot[3*nkpt-1]],(r"$\Gamma$","M","K",r"$\Gamma$"),fontsize=20)
    plt.yticks(fontsize=20)
    plt.grid(True)
    plt.xlim([kplot[0],kplot[-1]])
    
# Density of states
    
dE = 0.1
L = 100

ksamp = np.zeros([3,L**2])
Esamp = np.zeros([numbands,L**2])
count = 0
for i in range(L):
    for j in range(L):
        ksamp[:,count] = i/L*b1 + j/L*b2
        count = count + 1

for i in range(ksamp.shape[1]):
    
    if next_nearest == 1:
        Ha = tmatrix(ksamp[:,i],NNLIST,orbitals)
    else:
        Ha = np.diag([es,ep,ep,ep])
    Hab = tmatrix(ksamp[:,i],NLIST,orbitals)
    Hba = Hab.conj().transpose()
    Hb = Ha
    
    Hmat = np.kron( np.array([ [1,0], [0,0] ]), Ha) + np.kron( np.array([ [0,1], [0,0] ]), Hab)\
            + np.kron( np.array([ [0,0], [1,0] ]), Hba) + np.kron( np.array([ [0,0], [0,1] ]), Hb)
        
    e,v = np.linalg.eigh(Hmat)
    Esamp[:,i] = e
    
Emin = np.min(np.min(Esamp))
Emax = np.max(np.max(Esamp))
Esteps = np.arange(Emin,Emax,dE)
DOS = np.zeros([len(Esteps)])
for i in range(len(Esteps)-1): 
    for j in range(numbands):
        DOS[i] = DOS[i] + 1/dE * ( stateCount(Esamp[j,:],Esteps[i],Esteps[i+1]) )

g = plt.figure(figsize=(18, 10)); 
plt.plot(Esteps,DOS,color=cm)
plt.xlabel('Energy [eV]',fontsize=20)
plt.ylabel('# States/eV',fontsize=20)
plt.title('Density of States',fontsize=20)
plt.grid(True)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.axis('tight')
    
```
